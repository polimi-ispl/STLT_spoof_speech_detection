```python
import pandas as pd
from pycaret.classification import *
```

```python
nfft = 128
hop_size = 64 
```

```python
df_train = pd.read_pickle('../features/bicoherences/dataframes/train_bicoh_stats_nfft_{}_hop_size_{}'.format(
    nfft, hop_size))
```

```python
df_dev = pd.read_pickle('../features/bicoherences/dataframes/dev_bicoh_stats_nfft_{}_hop_size_{}'.format(
    nfft, hop_size))
```

```python
df_eval = pd.read_pickle('../features/bicoherences/dataframes/eval_bicoh_stats_nfft_{}_hop_size_{}'.format(
    nfft, hop_size))
```

```python
df_train
```

```python
import itertools
unknown_number = 2
unknown_label = 'unknown'
multiclass_list= {'-', 'A01', 'A02', 'A03', 'A04', 'A05', 'A06'}
unknown_combinations = itertools.combinations(multiclass_list, unknown_number)


for u in unknown_combinations:
    multiclass_dict = {'-':'bonafide', 'A01':'A01', 'A02':'A02', 'A03':'A03', 'A04':'A04', 'A05':'A05', 'A06':'A06'}
    
    if u[0] == '-' or u[1] == '-':
        continue
        
    # label 8 corresponds to unknown known
    for i in range(len(u)):
        multiclass_dict[u[i]] = unknown_label 
    
    df_train_open_set = df_train.replace({'system_id': multiclass_dict})
    df_dev_open_set = df_dev.replace({'system_id': multiclass_dict})
    df_eval_open_set = df_eval.copy()
    df_eval_open_set['system_id'] ='unknown'
    
    clf = setup(df_train_open_set, 
                target = 'system_id',
               ignore_features= ['audio_filename', 'speaker_id', 'label'],
               normalize=True,
               )
    print("Unknown classes are {} and {} for dev".format(u[0], u[1]))
    compare_models()
    
    
    
    
```
