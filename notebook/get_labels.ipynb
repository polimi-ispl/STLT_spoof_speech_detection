```python
import pandas as pd
import shutil
import os
```

## Get labels for train

```python
txt_train_path = '/nas/home/cborrelli/bot_speech/dataset/LA/ASVspoof2019_LA_cm_protocols/ASVspoof2019.LA.cm.train.trn.txt'
train_labels = pd.read_csv(txt_train_path, sep=" ", header=None)
train_labels.columns = ["SPEAKER_ID", "AUDIO_FILE_NAME", "-", "SYSTEM_ID", "KEY"]
```

```python
train_labels['SYSTEM_ID'].unique()
```

## Get labels for dev

```python
txt_dev_path = '/nas/home/cborrelli/bot_speech/dataset/LA/ASVspoof2019_LA_cm_protocols/ASVspoof2019.LA.cm.dev.trl.txt'
dev_labels = pd.read_csv(txt_dev_path, sep=" ", header=None)
dev_labels.columns = ["SPEAKER_ID", "AUDIO_FILE_NAME", "-", "SYSTEM_ID", "KEY"]
```

```python
dev_labels['SYSTEM_ID'].unique()
```

## Get labels for eval

```python
txt_eval_path = '/nas/home/cborrelli/bot_speech/dataset/LA/ASVspoof2019_LA_cm_protocols/ASVspoof2019.LA.cm.eval.trl.txt'
eval_labels = pd.read_csv(txt_eval_path, sep=" ", header=None)
eval_labels.columns = ["SPEAKER_ID", "AUDIO_FILE_NAME", "-", "SYSTEM_ID", "KEY"]
```

```python
eval_labels['SYSTEM_ID'].unique()
```

```python
eval_labels[eval_labels['KEY']=='spoof']
```

## Merge

```python

labels = pd.concat([train_labels, dev_labels, eval_labels], keys=['train', 'dev', 'eval'])
```
